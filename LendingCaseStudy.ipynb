#### Standardize data by removing special characters, or converting values for better analysis

```python
# Remove special characters from columns 'int_rate' and 'revol_util' to standardized data for analysis

df.int_rate = df.int_rate.apply(lambda x : float(x.replace('%','')))
df.revol_util = df.revol_util.apply(lambda x : float(x.replace('%','')))
```

```python
# Remove special characters from column 'emp_length' by changing '< 1 year' to 0 and '10+ years' to 10 and remove words years/year from the records

df.emp_length = pd.to_numeric(df.emp_length.apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
```

```python
# Remove word moths from term column
df.term = pd.to_numeric(df.term.apply(lambda x: int(x.replace(' months', ''))))
```

#### Define a function for Group and percentage calculation of default loan applications vs total loan applications

```python
def percent_charged_off (group_by_column):
    
    df1 = df.groupby(group_by_column).agg(total_applications = ("loan_status","count"),
    charged_off = ("loan_status",lambda x:(x == "Charged Off").sum()))
    
    df1["percentage_charged_off"]=(df1["charged_off"]/df1["total_applications"])*100
    
    return df1
```
